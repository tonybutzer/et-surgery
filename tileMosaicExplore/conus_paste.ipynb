```python
from Tile_to_AOI_mosaicker_funcs import xr_build_mosaic_ds
from Tile_to_AOI_mosaicker_funcs import xr_write_geotiff_from_ds
```

```python
# months = ['05','06','07','08','09','10','11','12']
months = ['07','08','09','10','11','12']
years = ['2020']
product ='etasw'
```

```python
! aws s3 ls ws-enduser/USA/
```

```python
tiles = ['USA/r50.0_tile7',]
conuss = ['USA/conus_mos',]

tile_list=[]
conus_list=[]
for year in years:
    for mon in months:
        tif_list = []
        for tile in tiles:
            mytile = tile+'/'+year+'/'+product+'_'+year+mon+'.tif'
            tile_list.append(mytile)
        for conus in conuss:    
            myconus = conus+'/'+year+'/'+product+'_'+year+mon+'.tif'
            conus_list.append(myconus)
```

```python
bucket = 'ws-enduser'
product = 'etasw'
out_prefix_path = bucket+'/USA/conus_mos/' # it gets the "year" folder from the function
```

```python
for i in range(0,len(months)):
    print((tile_list[i], conus_list[i]))
    mytifs = [tile_list[i], conus_list[i]]
    print(mytifs)
    tif_list = mytifs
    DS = xr_build_mosaic_ds(bucket, product, tif_list)
    primary_name = tif_list[0]  #first tif in list
    # outputs a geotiff (COG) from the mosaic in memory and places it into the S3 bucket
    xr_write_geotiff_from_ds(DS, primary_name, out_prefix_path)
```
